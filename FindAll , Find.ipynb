```python
import urllib
from bs4 import BeautifulSoup
api='https://twitter.com/realDonaldTrump?ref_src=twsrc%5Egoogle%7Ctwcamp%5Eserp%7Ctwgr%5Eauthor'
```

```python
def scraping(url):
    page=urllib.request.urlopen(url)
    soup=BeautifulSoup(page,'html.parser')
    print(soup.title)
scraping(api)
```

```python
def scraping(url):
    page=urllib.request.urlopen(url)
    soup=BeautifulSoup(page,'html.parser')
    for link in soup.findAll('a'):
        print(link.get('href'))
scraping(api)
```

```python
def scraping(url):
    page=urllib.request.urlopen(url)
    soup=BeautifulSoup(page,'html.parser')
    for link in soup.findAll('a'):
        print(link.text)
scraping(api)
```

```python
def scraping(url):
    page=urllib.request.urlopen(url)
    soup=BeautifulSoup(page,'html.parser')
    print(soup.find('div',{'class':'ProfileHeaderCard'}).find('a',{'class':'u-textUserColor'}).text)
scraping(api)
```

```python
def scraping(url):
    page=urllib.request.urlopen(url)
    soup=BeautifulSoup(page,'html.parser')
    i=1
    for link in soup.findAll('div',{'class':'content'}):
        print(i) 
        print(link.find('p').text)
        i+=1
scraping(api)
```
