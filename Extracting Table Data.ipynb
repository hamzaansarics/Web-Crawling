```python
import urllib
from bs4 import BeautifulSoup
```

```python
api='https://www.basketball-reference.com/players/a/'
```

```python
def scraping(url):
    page=urllib.request.urlopen(url)
    return BeautifulSoup(page,'html.parser')
soup=scraping(api)
for values in soup.findAll('tbody'):
    for val in values.findAll('tr'):
        print(val.text)
```

```python
for values in soup.findAll('tr'):
    for val in values.findAll(['td','th']):
        print(val.text+',')
```
